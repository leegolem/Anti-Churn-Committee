```python
df_raw01 = pd.read_csv(USER_PROFILE_PATH)
df_raw02 = pd.read_csv(EVENT_LOG_PATH)

print("df_raw01 shape:", df_raw01.shape)
print("df_raw02 shape:", df_raw02.shape)

display(df_raw01.head())
display(df_raw02.head())
```

```python
user_df = df_raw01.copy()
event_df = df_raw02.copy()
```

이벤트 타임 결측치 플레그 처리 및 알수 없음으로

```python
event_df["is_event_type_missing"] = (
    event_df["event_type"].isna()
)

event_df["event_type"] = (
    event_df["event_type"]
    .fillna("알수없음")
)
```

세션 아이디 전처리

```python
# 결측여부 표시
event_df["is_session_id_missing"] = (
    event_df["session_id"].isna()
)
```

notification_changed_date의 결측치는 변경이없음을 의미하므로 signup_date값을 채워넣음

```python
user_df["notification_changed_date"] = (
    user_df["notification_changed_date"]
    .fillna(user_df["signup_date"])
)
```

# 파생컬럼 생성

```python
type(event_df["event_time"])
```

사용자 정보랑 결합

```python
event_tableau_df = event_df.merge(
    user_df[
        [
            "user_id",
            "signup_date",
            "signup_month",
            "signup_channel",
            "device",
            "notification_agreed"
        ]
    ],
    on="user_id",
    how="left",
    validate="many_to_one"
)
```

가입후 경과일

```python
event_tableau_df["event_date"] = pd.to_datetime(
    event_tableau_df["event_date"],
    errors="coerce"
)

event_tableau_df["signup_date"] = pd.to_datetime(
    event_tableau_df["signup_date"],
    errors="coerce"
)

event_tableau_df["days_since_signup"] = (
    event_tableau_df["event_date"]
    - event_tableau_df["signup_date"]
).dt.days
```

```python
event_tableau_df["days_since_signup"] = (
    event_tableau_df["event_date"]
    - event_tableau_df["signup_date"]
).dt.days
```

활동 이벤트 분류

```python
#활동 이벤트를 구분
activity_events = [
    "앱실행",
    "온보딩_완료",
    "챌린지_탐색",
    "챌린지참여",
    "수면기록",
    "운동기록",
    "마음챙김",
    "식단기록",
    "알림오픈"
]
```

```python
event_tableau_df["is_user_activity"] = (
    event_tableau_df["event_type"]
    .isin(activity_events)
)
```

```python
#알수없음 제외
event_tableau_df["is_unknown_event"] = (
    event_tableau_df["event_type"] == "알수없음"
)
```

```python
# 세션 기준 분석용 플래그
event_tableau_df["is_valid_session_event"] = (
    event_tableau_df["session_id"].notna()
)
```

```python
event_tableau_df.head()
```

D3,D7,D30 리텐션 생성

```python
retention_activity_df = event_tableau_df[
    (event_tableau_df["is_valid_session_event"])
    & (event_tableau_df["is_user_activity"])
    & (event_tableau_df["days_since_signup"]>=0)
].copy()
```

# 사용자별 리텐션

```python
retention_days = [3, 7, 30]

user_retention_df = user_df[
    [
        "user_id",
        "signup_date",
        "signup_month",
        "signup_channel",
        "device",
        "notification_agreed"
    ]
].copy()
```

```python
user_retention_df["signup_date"] = pd.to_datetime(
    user_retention_df["signup_date"],
    errors="coerce"
)
```

```python
#데이터 종료일
data_end_date = event_tableau_df["event_date"].max()

user_retention_df["observable_days"] = (
    data_end_date
    - user_retention_df["signup_date"]
).dt.days
```

```python
# 리텐션 및 관찰 가능 여부
for day in retention_days:

    retained_users = (
        retention_activity_df.loc[
            retention_activity_df["days_since_signup"] == day,
            "user_id"
        ]
        .dropna()
        .unique()
    )

    user_retention_df[f"D{day}_retained"] = (
        user_retention_df["user_id"]
        .isin(retained_users)
    )

    user_retention_df[f"D{day}_observable"] = (
        user_retention_df["observable_days"] >= day
    )
```

# 저장

```python
TABLEAU_OUTPUT_DIR = DATA_TABLEAU_DIR

TABLEAU_OUTPUT_DIR.mkdir(
    parents=True,
    exist_ok=True
)


TABLEAU_EVENT_FILE = "tableau_event_detail.csv"
TABLEAU_USER_FILE = "tableau_user_retention.csv"

TABLEAU_EVENT_PATH = TABLEAU_OUTPUT_DIR/ TABLEAU_EVENT_FILE

TABLEAU_USER_PATH = TABLEAU_OUTPUT_DIR/ TABLEAU_USER_FILE
```

```python
event_tableau_df.to_csv(TABLEAU_EVENT_PATH,index=False,encoding="utf-8-sig",date_format="%Y-%m-%d %H:%M:%S")
user_retention_df.to_csv(TABLEAU_USER_PATH,index=False,encoding="utf-8-sig",date_format="%Y-%m-%d")
print("Tableau 이벤트 상세 파일 저장 완료")
print("경로:", TABLEAU_EVENT_PATH)
print("데이터 크기:", event_tableau_df.shape)

print()

print("Tableau 사용자 리텐션 파일 저장 완료")
print("경로:", TABLEAU_USER_PATH)
print("데이터 크기:", user_retention_df.shape)
```
