# 기준 1: Day 0 기준 설정

리텐션 분석에서 Day 0은 사용자의 기준 시작일을 의미한다.  
Day 0을 어디로 설정하느냐에 따라 리텐션 분석의 해석 범위가 달라진다.

이번 분석에서는 아래 4가지 후보를 비교한다.

| Day 0 후보 | 의미 | 장점 | 주의점 |
|---|---|---|---|
| 가입일 | 사용자가 서비스에 가입한 날짜 | 전체 신규 가입자 기준으로 리텐션을 볼 수 있음 | 가입만 하고 앱을 실행하지 않은 사용자도 포함됨 |
| 첫 앱실행일 | 사용자가 처음 앱을 실행한 날짜 | 실제 앱 사용 시작 이후의 리텐션을 볼 수 있음 | 가입 후 앱을 실행하지 않은 사용자가 제외됨 |
| 온보딩 완료일 | 사용자가 온보딩을 완료한 날짜 | 온보딩 이후 유지율을 보기 좋음 | 온보딩을 완료하지 못한 사용자가 제외되어 리텐션이 높게 보일 수 있음 |
| 첫 핵심기능 사용일 | 기록/챌린지 등 핵심 기능을 처음 사용한 날짜 | 기능 경험 이후 잔존을 보기 좋음 | 적극적인 사용자만 남아 선택 편향이 커질 수 있음 |

따라서 각 후보별로 기준일을 만들 수 있는 사용자 수와 가입일 이후 기준일까지 걸린 시간을 비교한 뒤,  
이번 과제의 목적에 가장 적합한 Day 0 기준을 선택한다.

## 1-1. 가입월별 사용자 수 확인

```python
# Day 0 후보군 비교

# 후보 1. 가입일
day0_signup = user_profile[["user_id", "signup_day"]].copy()
day0_signup = day0_signup.rename(columns={"signup_day": "day0_date"})
day0_signup["day0_type"] = "가입일"

# 후보 2. 첫 앱실행일
first_app_launch = (
    event_log[event_log["event_type"].eq("앱실행")]
    .groupby("user_id", as_index=False)
    .agg(day0_date=("event_date", "min"))
)
first_app_launch["day0_type"] = "첫 앱실행일"

# 후보 3. 온보딩 완료일
first_onboarding = (
    event_log[event_log["event_type"].eq("온보딩_완료")]
    .groupby("user_id", as_index=False)
    .agg(day0_date=("event_date", "min"))
)
first_onboarding["day0_type"] = "온보딩 완료일"

# 후보 4. 첫 핵심기능 사용일
core_events = [
    "수면기록",
    "운동기록",
    "마음챙김",
    "식단기록",
    "챌린지_탐색",
    "챌린지참여"
]

first_core_event = (
    event_log[event_log["event_type"].isin(core_events)]
    .groupby("user_id", as_index=False)
    .agg(day0_date=("event_date", "min"))
)
first_core_event["day0_type"] = "첫 핵심기능 사용일"

# 후보군 통합
day0_candidates = pd.concat(
    [
        day0_signup,
        first_app_launch,
        first_onboarding,
        first_core_event
    ],
    ignore_index=True
)

# 가입일과 비교하기 위해 signup_day 결합
day0_candidates = day0_candidates.merge(
    user_profile[["user_id", "signup_day"]],
    on="user_id",
    how="left"
)

# 가입 후 기준일까지 걸린 일수
day0_candidates["days_from_signup"] = (
    day0_candidates["day0_date"] - day0_candidates["signup_day"]
).dt.days

display(day0_candidates.head())
```

```python
# Day 0 후보별 적용 가능 사용자 수 비교

total_user_count = user_profile["user_id"].nunique()

day0_candidate_summary = (
    day0_candidates
    .groupby("day0_type", as_index=False)
    .agg(
        기준일_생성_사용자수=("user_id", "nunique"),
        가입후_기준일까지_평균일수=("days_from_signup", "mean"),
        가입후_기준일까지_중앙값=("days_from_signup", "median"),
        가입후_기준일까지_최대일수=("days_from_signup", "max")
    )
)

day0_candidate_summary["전체_가입자수"] = total_user_count

day0_candidate_summary["기준일_없는_사용자수"] = (
    day0_candidate_summary["전체_가입자수"]
    - day0_candidate_summary["기준일_생성_사용자수"]
)

day0_candidate_summary["기준일_생성률"] = (
    day0_candidate_summary["기준일_생성_사용자수"]
    / day0_candidate_summary["전체_가입자수"]
    * 100
).round(2)

day0_candidate_summary["가입후_기준일까지_평균일수"] = (
    day0_candidate_summary["가입후_기준일까지_평균일수"].round(2)
)

day0_type_order = [
    "가입일",
    "첫 앱실행일",
    "온보딩 완료일",
    "첫 핵심기능 사용일"
]

day0_candidate_summary["day0_type"] = pd.Categorical(
    day0_candidate_summary["day0_type"],
    categories=day0_type_order,
    ordered=True
)

day0_candidate_summary = (
    day0_candidate_summary
    .sort_values("day0_type")
    [[
        "day0_type",
        "전체_가입자수",
        "기준일_생성_사용자수",
        "기준일_없는_사용자수",
        "기준일_생성률",
        "가입후_기준일까지_평균일수",
        "가입후_기준일까지_중앙값",
        "가입후_기준일까지_최대일수"
    ]]
)

display(day0_candidate_summary)
```

### 기준일 생성률

```python
plt.figure(figsize=(9, 5))

bars = plt.bar(
    day0_candidate_summary["day0_type"].astype(str),
    day0_candidate_summary["기준일_생성률"]
)

plt.title("Day 0 후보별 기준일 생성률")
plt.xlabel("Day 0 후보")
plt.ylabel("기준일 생성률(%)")
plt.ylim(0, 105)
plt.grid(axis="y", alpha=0.3)

plt.bar_label(
    bars,
    labels=[f"{value:.1f}%" for value in day0_candidate_summary["기준일_생성률"]],
    padding=3
)

plt.show()
```

### 그래프 해석: Day 0 후보별 기준일 생성률

가입일은 전체 사용자에게 적용 가능하다.     
첫 앱실행일과 첫 핵심기능 사용일도 대부분의 사용자에게 기준일을 만들 수 있다.   
온보딩 완료일은 기준일 생성률이 낮아, 온보딩을 완료하지 못한 사용자가 분석에서 제외

# 기준 2: 활성 유저 판단 이벤트 설정
리텐션에서 활성 유저는 **해당 일자에 실제로 앱에 다시 들어온 사용자**를 의미한다.       
따라서 이번 분석에서는 `앱실행` 이벤트가 존재하는 사용자를 활성 유저로 판단한다.


리텐션은 사용자가 앱에 다시 돌아왔는지를 보는 지표로써      
기록, 챌린지, 알림 이벤트 보단, 실제 앱 진입을 의미하는 `앱실행`을 활성 판단 기준으로 두는것이 적절

## Event Type별 이벤트 수 확인
어떤 이벤트가 얼마나 발생했는지 확인한 뒤, 리텐션 활성 판단에 사용할 이벤트와 제외할 이벤트를 구분

```python
# 
event_type_summary = (
    event_log
    .assign(event_type_display=event_log["event_type"].fillna("Event_Type 결측"))
    .groupby("event_type_display", as_index=False)
    .agg(
        event_count=("user_id", "size"),
        user_count=("user_id", "nunique")
    )
    .sort_values("event_count", ascending=False)
)

event_type_summary["retention_active_judgement"] = np.where(
    event_type_summary["event_type_display"].eq("앱실행"),
    "활성 인정",
    "활성 제외"
)

event_type_summary["exclude_reason"] = np.select(
    [
        event_type_summary["event_type_display"].eq("앱실행"),
        event_type_summary["event_type_display"].isin(["알림수신", "알림오픈"]),
        event_type_summary["event_type_display"].eq("Event_Type 결측")
    ],
    [
        "앱을 실제로 실행한 이벤트",
        "앱 밖 OS 알림 이벤트",
        "식별 불가능한 이벤트"
    ],
    default="기능 사용 이벤트이지만 활성 판단 기준에서는 제외"
)

display(event_type_summary)
```

### 그래프 해석: Event_Type별 이벤트 수
이벤트 유형별 발생량을 비교

`앱실행` 이벤트는 사용자가 실제로 앱에 진입했다는 신호이므로, 리텐션 활성 판단 기준으로 사용

따라서 리텐션 계산에서는 `앱실행`만 활성 이벤트로 사용하고, 나머지 이벤트는 이후 원인 분석에서 활용한다.

## 3-2. 알림 이벤트의 Session_ID 결측 확인

과제 설명상 알림 이벤트는 앱 밖에서 발생하므로 `Session_ID`가 없는 것이 정상        
따라서 알림 이벤트의 `Session_ID` 결측률을 확인해, 알림 이벤트를 앱 내부 활동으로 보기 어려운지 점검한다.

```python
notification_session_summary = (
    event_log
    .assign(
        is_notification_event=event_log["event_type"].isin(notification_events),
        is_session_missing=event_log["session_id"].isna()
    )
    .groupby("is_notification_event", as_index=False)
    .agg(
        event_count=("user_id", "size"),
        user_count=("user_id", "nunique"),
        session_missing_count=("is_session_missing", "sum")
    )
)

notification_session_summary["session_missing_ratio"] = (
    notification_session_summary["session_missing_count"]
    / notification_session_summary["event_count"]
    * 100
).round(2)

notification_session_summary["event_group"] = np.where(
    notification_session_summary["is_notification_event"],
    "알림 이벤트",
    "그 외 이벤트"
)

notification_session_summary = notification_session_summary[[
    "event_group",
    "event_count",
    "user_count",
    "session_missing_count",
    "session_missing_ratio"
]]

display(notification_session_summary)

plt.figure(figsize=(7, 5))

bars = plt.bar(
    notification_session_summary["event_group"],
    notification_session_summary["session_missing_ratio"]
)

plt.title("알림 이벤트 여부별 Session_ID 결측률")
plt.xlabel("이벤트 구분")
plt.ylabel("Session_ID 결측률(%)")
plt.ylim(0, 105)
plt.grid(axis="y", alpha=0.3)

plt.bar_label(
    bars,
    labels=[f"{value:.1f}%" for value in notification_session_summary["session_missing_ratio"]],
    padding=3
)

plt.show()
```

### 그래프 해석: Session_ID 결측률

알림 이벤트는 `Session_ID` 결측률이 높게 나타난다.      
이는 알림 이벤트가 앱 내부 세션에서 발생한 행동이 아니라, 앱 밖 OS 알림 환경에서 발생한 이벤트라는 설명과 일치한다.

따라서 `알림수신`, `알림오픈`은 리텐션 활성 판단에서는 제외하는 것이 적절하다.

## 4-1. 리텐션 분석용 이벤트 기준 테이블 생성

리텐션 기준 확인을 위해 이벤트 로그에 사용자 프로필 정보를 결합한다.  
여기서는 기존 전처리에서 만든 `is_log_issue_period`를 다시 만들지 않고 그대로 사용한다.

추가로 아래 컬럼만 생성한다.

| 컬럼 | 의미 |
|---|---|
| `event_day_from_signup` | 가입일 기준 이벤트 발생 경과일 |
| `is_retention_active_event` | 리텐션 활성 판단에 사용할 `앱실행` 여부 |
| `is_notification_event` | 알림 이벤트 여부 |
| `is_event_before_signup` | 가입일 이전 이벤트 여부 |


```python
profile_for_merge = user_profile[[
    "user_id",
    "signup_day",
    "signup_month",
    "is_no_event_user",
    "is_signup_log_issue_period"
]].copy()

event_log_retention_base = event_log.merge(
    profile_for_merge,
    on="user_id",
    how="left"
)

event_log_retention_base["event_day_from_signup"] = (
    event_log_retention_base["event_date"] - event_log_retention_base["signup_day"]
).dt.days

event_log_retention_base["is_retention_active_event"] = (
    event_log_retention_base["event_type"].eq("앱실행")
)

event_log_retention_base["is_notification_event"] = (
    event_log_retention_base["event_type"].isin(["알림수신", "알림오픈"])
)

event_log_retention_base["is_event_before_signup"] = (
    event_log_retention_base["event_day_from_signup"].lt(0)
)

retention_base_check = pd.DataFrame({
    "항목": [
        "이벤트 로그 행 수",
        "프로필과 매칭되지 않은 이벤트 수",
        "가입일 이전 이벤트 수",
        "리텐션 활성 이벤트 수",
        "알림 이벤트 수",
        "로그 장애 기간 이벤트 수"
    ],
    "값": [
        len(event_log_retention_base),
        int(event_log_retention_base["signup_day"].isna().sum()),
        int(event_log_retention_base["is_event_before_signup"].sum()),
        int(event_log_retention_base["is_retention_active_event"].sum()),
        int(event_log_retention_base["is_notification_event"].sum()),
        int(event_log_retention_base["is_log_issue_period"].sum())
    ]
})

display(retention_base_check)
display(event_log_retention_base.head())
```

### 표 해석: 리텐션 분석용 이벤트 기준 테이블

리텐션 분석에 사용할 기준 컬럼이 정상적으로 만들어졌는지 확인        
특히 `is_log_issue_period`는 새로 생성한 값이 아니라, 전처리 완료 데이터에 있는 기존 플래그를 사용

이후 리텐션 계산에서는 `is_retention_active_event`가 True인 `앱실행` 이벤트만 활성 판단에 사용

## 기준 결정
로그 수집 장애 기간은 삭제하지 않고 **플래그로 관리**한다.

- 장애 기간: 2025-03-10 ~ 2025-03-14
- 이벤트 기준 플래그: `is_log_issue_period`
- 처리 방식: 삭제하지 않고 해석 주의 구간으로 표시
- 이후 리텐션 분석: D0/D3/D7/D30 목표일이 장애 기간에 걸리는 사용자를 별도 플래그로 확인

즉, 장애 기간에 활동이 없다고 해서 무조건 이탈로 단정하지 않고, 데이터 품질 이슈 가능성을 함께 고려한다.

# 기준 5: D0/D3/D7/D30 목표일 산정 가능 여부 확인
리텐션을 계산하려면 각 사용자별로 목표일이 필요하다.  
예를 들어 D7 리텐션은 `가입일 + 7일`에 해당하는 날짜에 앱실행 이벤트가 있는지 확인하는 방식이다.

다만 목표일이 이벤트 관측 기간 밖에 있거나, 로그 수집 장애 기간과 겹치면 리텐션 해석에 주의가 필요함

## 사용자별 리텐션 목표일 생성

```python
# 사용자별 D0/D3/D7/D30 목표일 생성

retention_day_df = pd.DataFrame({
    "retention_day": retention_days_list
})

user_retention_targets = (
    user_profile[[
        "user_id",
        "signup_day",
        "signup_month",
        "is_no_event_user",
        "is_signup_log_issue_period"
    ]]
    .merge(retention_day_df, how="cross")
)

user_retention_targets["target_date"] = (
    user_retention_targets["signup_day"]
    + pd.to_timedelta(user_retention_targets["retention_day"], unit="D")
)

user_retention_targets["retention_label"] = (
    "D" + user_retention_targets["retention_day"].astype(str)
)

event_observation_start = event_log["event_date"].min()
event_observation_end = event_log["event_date"].max()

user_retention_targets["is_target_observable"] = (
    user_retention_targets["target_date"].between(
        event_observation_start,
        event_observation_end
    )
)

user_retention_targets["is_target_log_issue_period"] = (
    user_retention_targets["target_date"].between(
        issue_start,
        issue_end
    )
)

display(user_retention_targets.head())
```

## 목표일 산정 가능 여부 요약표

```python
target_observable_summary = (
    user_retention_targets
    .groupby(["retention_day", "retention_label"], as_index=False)
    .agg(
        target_user_count=("user_id", "nunique"),
        observable_user_count=("is_target_observable", "sum"),
        log_issue_target_user_count=("is_target_log_issue_period", "sum")
    )
    .sort_values("retention_day")
)

target_observable_summary["observable_rate"] = (
    target_observable_summary["observable_user_count"]
    / target_observable_summary["target_user_count"]
    * 100
).round(2)

target_observable_summary["log_issue_target_rate"] = (
    target_observable_summary["log_issue_target_user_count"]
    / target_observable_summary["target_user_count"]
    * 100
).round(2)

target_observable_summary = target_observable_summary[[
    "retention_label",
    "target_user_count",
    "observable_user_count",
    "observable_rate",
    "log_issue_target_user_count",
    "log_issue_target_rate"
]]

display(target_observable_summary)
```

# 기준 6: 월별 가입 코호트 기준 분석 가능 범위 확인

월별 가입자 집단마다 리텐션 목표일이 로그 장애 기간과 겹치는 정도가 다를 수 있다.  

## 6-1. 가입월 × 리텐션 목표일별 산정 가능 상태 확인

각 가입월 코호트에 대해 D0/D3/D7/D30 목표일이 정상적으로 산정 가능한지 확인

| 상태 | 의미 |
|---|---|
| 정상 산정 가능 | 목표일이 관측 기간 안에 있고, 로그 장애 기간과 겹치지 않음 |
| 장애기간 중복 | 목표일이 로그 수집 장애 기간과 겹침 |
| 관측 불가 | 목표일이 이벤트 관측 기간 밖에 있음 |


```python
# 가입월 × 리텐션 목표일별 산정 가능 상태 확인

user_retention_targets["retention_target_status"] = np.select(
    [
        ~user_retention_targets["is_target_observable"],
        user_retention_targets["is_target_log_issue_period"],
        user_retention_targets["is_target_observable"] & ~user_retention_targets["is_target_log_issue_period"]
    ],
    [
        "관측 불가",
        "장애기간 중복",
        "정상 산정 가능"
    ],
    default="확인 필요"
)

cohort_target_summary = (
    user_retention_targets
    .groupby(["signup_month", "retention_day", "retention_label"], as_index=False)
    .agg(
        cohort_user_count=("user_id", "nunique"),
        observable_user_count=("is_target_observable", "sum"),
        log_issue_target_user_count=("is_target_log_issue_period", "sum")
    )
)

cohort_target_summary["normal_target_user_count"] = (
    cohort_target_summary["observable_user_count"]
    - cohort_target_summary["log_issue_target_user_count"]
)

cohort_target_summary["observable_rate"] = (
    cohort_target_summary["observable_user_count"]
    / cohort_target_summary["cohort_user_count"]
    * 100
).round(2)

cohort_target_summary["normal_target_rate"] = (
    cohort_target_summary["normal_target_user_count"]
    / cohort_target_summary["cohort_user_count"]
    * 100
).round(2)

cohort_target_summary["log_issue_target_rate"] = (
    cohort_target_summary["log_issue_target_user_count"]
    / cohort_target_summary["cohort_user_count"]
    * 100
).round(2)

cohort_target_summary["retention_label"] = pd.Categorical(
    cohort_target_summary["retention_label"],
    categories=["D0", "D3", "D7", "D30"],
    ordered=True
)

cohort_target_summary = cohort_target_summary.sort_values(
    ["signup_month", "retention_label"]
)

display(cohort_target_summary)
```

## 6-2. 가입월 × 리텐션 목표일 정상 산정 가능률 히트맵

```python
normal_rate_matrix = cohort_target_summary.pivot(
    index="signup_month",
    columns="retention_label",
    values="normal_target_rate"
)

normal_rate_matrix = normal_rate_matrix[["D0", "D3", "D7", "D30"]]

display(normal_rate_matrix)

plt.figure(figsize=(8, 5))
plt.imshow(normal_rate_matrix, aspect="auto")

plt.title("가입월별 리텐션 목표일 정상 산정 가능률")
plt.xlabel("리텐션 목표일")
plt.ylabel("가입월")

plt.xticks(
    ticks=np.arange(len(normal_rate_matrix.columns)),
    labels=normal_rate_matrix.columns
)

plt.yticks(
    ticks=np.arange(len(normal_rate_matrix.index)),
    labels=normal_rate_matrix.index
)

for row_idx in range(normal_rate_matrix.shape[0]):
    for col_idx in range(normal_rate_matrix.shape[1]):
        value = normal_rate_matrix.iloc[row_idx, col_idx]
        plt.text(
            col_idx,
            row_idx,
            f"{value:.1f}%",
            ha="center",
            va="center"
        )

plt.colorbar(label="정상 산정 가능률(%)")
plt.show()
```

### 그래프 해석: 가입월별 정상 산정 가능률
대부분의 가입월과 목표일은 정상 산정 가능률이 높아 리텐션 비교에 사용할 수 있다.

# 추가 EDA: 첫 7일 이내 재방문 여부와 D30 리텐션 비교

조원이 제안한 추가 가설을 확인한다.

> 7일 이내 재방문하지 않은 유저는 D30 리텐션이 유의하게 낮을 것이다.

이 분석은 위에서 정리한 리텐션 기준을 새로 바꾸기 위한 파트가 아니라, 최종 기준을 정리한 뒤 추가로 확인하는 EDA다.

| 구분 | 기준 |
|---|---|
| Day 0 | 가입일 |
| 재방문 기준 | 가입 후 D1~D7 사이 `앱실행` 1회 이상 |
| D30 리텐션 기준 | 가입일 + 30일째 되는 날짜에 `앱실행` 이벤트 존재 |
| 주 분석 대상 | 첫 7일 관측 구간 또는 D30 목표일이 로그 장애 기간과 겹치지 않는 사용자 |
| 보조 분석 | D1~D3까지만 재방문한 사용자와 D4~D7에도 재방문한 사용자의 D30 리텐션 비교 |


## 1. 첫 7일 재방문 / D30 리텐션 기준 날짜 생성

사용자별로 첫 7일 관측 구간과 D30 목표일을 만든다.


```python
revisit_base = user_profile[[
    "user_id",
    "signup_day",
    "signup_month"
]].copy()

revisit_base["first7_start_date"] = revisit_base["signup_day"] + pd.Timedelta(days=1)
revisit_base["first7_end_date"] = revisit_base["signup_day"] + pd.Timedelta(days=7)
revisit_base["d30_date"] = revisit_base["signup_day"] + pd.Timedelta(days=30)

revisit_base["is_first7_window_log_issue_period"] = (
    (revisit_base["first7_start_date"] <= issue_end)
    & (revisit_base["first7_end_date"] >= issue_start)
)

revisit_base["is_d30_target_log_issue_period"] = revisit_base["d30_date"].between(
    issue_start,
    issue_end
)

display(revisit_base.head())
```

## 2. 앱실행 로그 기준 재방문 여부와 D30 잔존 여부 생성

`앱실행` 이벤트만 사용해서 첫 7일 이내 재방문 여부, 첫 주 전반/후반 재방문 단계, D30 리텐션 여부를 만든다.


```python
app_launch_log = (
    event_log[event_log["event_type"] == "앱실행"]
    [["user_id", "event_date"]]
    .drop_duplicates()
)

first7_app_launch = app_launch_log.merge(
    revisit_base[["user_id", "signup_day", "first7_start_date", "first7_end_date"]],
    on="user_id",
    how="left"
)

first7_app_launch = first7_app_launch[
    (first7_app_launch["event_date"] >= first7_app_launch["first7_start_date"])
    & (first7_app_launch["event_date"] <= first7_app_launch["first7_end_date"])
].copy()

first7_app_launch["revisit_day"] = (
    first7_app_launch["event_date"] - first7_app_launch["signup_day"]
).dt.days

first7_revisited_users = (
    first7_app_launch["user_id"]
    .drop_duplicates()
)

early_revisited_users = (
    first7_app_launch[first7_app_launch["revisit_day"].between(1, 3)]["user_id"]
    .drop_duplicates()
)

late_revisited_users = (
    first7_app_launch[first7_app_launch["revisit_day"].between(4, 7)]["user_id"]
    .drop_duplicates()
)

d30_app_launch = app_launch_log.merge(
    revisit_base[["user_id", "d30_date"]],
    on="user_id",
    how="left"
)

d30_retained_users = (
    d30_app_launch[
        d30_app_launch["event_date"] == d30_app_launch["d30_date"]
    ]["user_id"]
    .drop_duplicates()
)

revisit_base["is_first7_revisited"] = revisit_base["user_id"].isin(first7_revisited_users)
revisit_base["is_early_revisited"] = revisit_base["user_id"].isin(early_revisited_users)
revisit_base["is_late_revisited"] = revisit_base["user_id"].isin(late_revisited_users)
revisit_base["is_d30_retained"] = revisit_base["user_id"].isin(d30_retained_users)

revisit_base["first7_revisit_group"] = np.where(
    revisit_base["is_first7_revisited"],
    "7일 이내 재방문",
    "7일 이내 재방문 없음"
)

revisit_base["first7_revisit_stage"] = np.select(
    [
        ~revisit_base["is_early_revisited"] & ~revisit_base["is_late_revisited"],
        revisit_base["is_early_revisited"] & ~revisit_base["is_late_revisited"],
        revisit_base["is_late_revisited"]
    ],
    [
        "7일 이내 재방문 없음",
        "D1~D3까지만 재방문",
        "D4~D7에도 재방문"
    ],
    default="확인 필요"
)

display(revisit_base.head())
```

## 3. 주 분석 대상 생성

첫 7일 관측 구간이나 D30 목표일이 로그 장애 기간과 겹치면 재방문 여부 또는 D30 리텐션이 과소 측정될 수 있다.  
따라서 주 분석에서는 해당 사용자를 제외하고 비교한다.


```python
revisit_analysis_base = revisit_base[
    (~revisit_base["is_first7_window_log_issue_period"])
    & (~revisit_base["is_d30_target_log_issue_period"])
].copy()

revisit_issue_summary = pd.DataFrame({
    "구분": [
        "전체 사용자",
        "첫 7일 관측 구간이 로그 장애 기간과 겹친 사용자",
        "D30 목표일이 로그 장애 기간과 겹친 사용자",
        "주 분석 대상 사용자"
    ],
    "사용자수": [
        len(revisit_base),
        revisit_base["is_first7_window_log_issue_period"].sum(),
        revisit_base["is_d30_target_log_issue_period"].sum(),
        len(revisit_analysis_base)
    ]
})

revisit_issue_summary["비율"] = (
    revisit_issue_summary["사용자수"] / len(revisit_base) * 100
).round(2)

display(revisit_issue_summary)
```

## 4. 첫 7일 이내 재방문 여부별 D30 리텐션

첫 7일 이내 재방문 그룹과 재방문 없음 그룹의 D30 리텐션을 비교한다.


```python
first7_summary = (
    revisit_analysis_base
    .groupby("first7_revisit_group")
    .agg(
        user_count=("user_id", "count"),
        d30_retained_count=("is_d30_retained", "sum")
    )
    .reset_index()
)

first7_summary["d30_retention_rate"] = (
    first7_summary["d30_retained_count"] / first7_summary["user_count"]
).round(4)

first7_summary["d30_retention_rate_percent"] = (
    first7_summary["d30_retention_rate"] * 100
).round(2)

first7_summary["sort_order"] = first7_summary["first7_revisit_group"].map({
    "7일 이내 재방문": 1,
    "7일 이내 재방문 없음": 2
})

first7_summary = (
    first7_summary
    .sort_values("sort_order")
    .drop(columns="sort_order")
)

display(first7_summary)
```

```python
first7_plot = (
    first7_summary
    .set_index("first7_revisit_group")
    .loc[["7일 이내 재방문", "7일 이내 재방문 없음"], "d30_retention_rate_percent"]
)

plt.figure(figsize=(7, 5))
bars = plt.bar(first7_plot.index, first7_plot.values)
plt.title("첫 7일 이내 재방문 여부별 D30 리텐션")
plt.xlabel("재방문 그룹")
plt.ylabel("D30 리텐션(%)")
plt.grid(axis="y", alpha=0.3)

for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f"{bar.get_height():.2f}%",
        ha="center",
        va="bottom"
    )

plt.tight_layout()
plt.show()
```

### 그래프 해석: 첫 7일 이내 재방문 여부별 D30 리텐션

첫 7일 이내 재방문한 사용자는 D30 리텐션이 높게 나타난다.  
반대로 첫 7일 이내 재방문하지 않은 사용자는 D30 리텐션이 매우 낮게 나타난다.

즉, 가입 직후 Day 0에서 앱을 실행한 것만으로는 충분하지 않고, 가입 후 첫 주 안에 다시 앱으로 돌아오는 행동이 장기 잔존과 강하게 연결되는 것으로 볼 수 있다.


## 5. 두 비율 z-test

첫 7일 이내 재방문 그룹과 재방문 없음 그룹의 D30 리텐션 차이가 통계적으로 유의한지 확인한다.

`Statistical_Test.ipynb`에서 사용한 방식처럼, 두 그룹을 합친 비율을 기준으로 표준오차를 계산하고 양측검정으로 p-value를 확인한다.


```python

# Statistical_Test.ipynb의 두 비율 z-test 계산 방식 참조
from scipy import stats

ALPHA = 0.05

first7_test_table = first7_summary.set_index("first7_revisit_group")

group_a = "7일 이내 재방문"
group_b = "7일 이내 재방문 없음"

success_a = first7_test_table.loc[group_a, "d30_retained_count"]
total_a = first7_test_table.loc[group_a, "user_count"]
success_b = first7_test_table.loc[group_b, "d30_retained_count"]
total_b = first7_test_table.loc[group_b, "user_count"]

rate_a = success_a / total_a
rate_b = success_b / total_b

# 귀무가설: 두 그룹의 D30 리텐션율은 같다
# z-test에서는 귀무가설 기준으로 두 그룹을 합친 비율 사용
pooled_rate = (success_a + success_b) / (total_a + total_b)

# 두 비율 차이의 표준오차
standard_error = np.sqrt(
    pooled_rate * (1 - pooled_rate) * (1 / total_a + 1 / total_b)
)

# z 검정통계량과 양측 p-value
z_stat = (rate_a - rate_b) / standard_error
p_value = 2 * stats.norm.sf(abs(z_stat))

# 리텐션율 차이(%p)
diff_pp = (rate_a - rate_b) * 100

first7_z_test = pd.DataFrame({
    "재방문_D30리텐션": [rate_a * 100],
    "재방문없음_D30리텐션": [rate_b * 100],
    "리텐션차이": [diff_pp],
    "z_stat": [z_stat],
    "p_value": [p_value],
    "유의여부": [p_value < ALPHA]
})

first7_z_test = first7_z_test.round(4)

display(first7_z_test)
```

### 결과 해석

- 첫 7일 이내 재방문한 사용자의 D30 리텐션은 29.90%
- 첫 7일 이내 재방문하지 않은 사용자의 D30 리텐션은 0.00% 

두 그룹의 리텐션 차이는 29.90%p였으며, 두 비율 z-test 결과 p-value가 0.001보다 작아 통계적으로 유의한 차이가 확인되었다.

따라서 가입 후 첫 7일 이내 재방문 여부는 D30 리텐션과 강하게 연결되어 있으며,       
초기 7일 안에 사용자를 다시 앱으로 유도하는 것이 리텐션 개선의 핵심 지점으로 볼 수 있다.


## 6. 보조 분석: 첫 주 전반/후반 재방문 단계별 D30 리텐션

앞서 D1~D7 사이에 한 번이라도 앱을 실행하면 모두 "첫 7일 이내 재방문"으로 분류했다.
이 분석만으론 가입 직후 하루 이틀만 다시 들어온 사용자와, 첫 주 후반까지 앱 사용이 이어진 사용자를 구분하기 어렵다.

따라서 이번 분석에서는 다음과 같이 세분화 해 재방문이 가입 초반에 그쳤는지 또는 첫 주 후반까지 이어졌는지 확인한다.
이를 통해 D30 리텐션과 더 강하게 연결되는 것이 단순 재방문 여부인지, 아니면 첫 주 후반까지 이어지는 반복 사용인지 살펴본다.

따라서 보조 분석에서는 **첫 7일 이내 재방문 사용자를 더 세분화하여, 재방문이 가입 초반에 그쳤는지 또는 첫 주 후반까지 이어졌는지 확인**한다.
이를 통해 D30 리텐션과 더 강하게 연결되는 것이 단순 재방문 여부인지, 아니면 첫 주 후반까지 이어지는 반복 사용인지 살펴본다.

| 그룹 | 기준 |
|---|---|
| 7일 이내 재방문 없음 | D1~D7 사이 앱실행 없음 |
| D1~D3까지만 재방문 | D1~D3에는 앱실행이 있으나 D4~D7에는 앱실행 없음 |
| D4~D7에도 재방문 | D4~D7에도 앱실행 있음 |

즉, 이 분석은 단순히 첫 7일 안에 다시 들어왔는지가 아니라, **첫 주 후반까지 사용이 이어진 사용자가 D30까지 더 잘 남는지**를 확인하기 위한 보조 분석이다.

```python
first7_stage_summary = (
    revisit_analysis_base
    .groupby("first7_revisit_stage")
    .agg(
        user_count=("user_id", "count"),
        d30_retained_count=("is_d30_retained", "sum")
    )
    .reset_index()
)

first7_stage_summary["d30_retention_rate"] = (
    first7_stage_summary["d30_retained_count"] / first7_stage_summary["user_count"]
).round(4)

first7_stage_summary["d30_retention_rate_percent"] = (
    first7_stage_summary["d30_retention_rate"] * 100
).round(2)

first7_stage_summary["sort_order"] = first7_stage_summary["first7_revisit_stage"].map({
    "7일 이내 재방문 없음": 1,
    "D1~D3까지만 재방문": 2,
    "D4~D7에도 재방문": 3
})

first7_stage_summary = (
    first7_stage_summary
    .sort_values("sort_order")
    .drop(columns="sort_order")
)

display(first7_stage_summary)
```

```python
first7_stage_plot = (
    first7_stage_summary
    .set_index("first7_revisit_stage")
    .loc[
        ["7일 이내 재방문 없음", "D1~D3까지만 재방문", "D4~D7에도 재방문"],
        "d30_retention_rate_percent"
    ]
)

plt.figure(figsize=(8, 5))
bars = plt.bar(first7_stage_plot.index, first7_stage_plot.values)
plt.title("첫 주 전반/후반 재방문 단계별 D30 리텐션")
plt.xlabel("재방문 단계")
plt.ylabel("D30 리텐션(%)")
plt.grid(axis="y", alpha=0.3)

for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f"{bar.get_height():.2f}%",
        ha="center",
        va="bottom"
    )

plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

### 그래프 해석: 첫 주 전반/후반 재방문 단계별 D30 리텐션

첫 주 전반인 D1~D3까지만 재방문한 사용자의 D30 리텐션은 0.00%로, 7일 이내 재방문이 없는 사용자와 동일하게 나타났다.

반면 D4~D7에도 재방문한 사용자의 D30 리텐션은 37.35%로 크게 높아졌다.  
즉, 가입 직후 한두 번 다시 들어오는 것보다 **첫 주 후반까지 재방문이 이어지는지**가 D30 잔존과 더 강하게 연결된다.

따라서 첫 주 리텐션 전략은 단순히 가입 직후 재방문을 한 번 만드는 것에서 끝나기보다, D4~D7 구간까지 앱 실행을 이어가도록 설계하는 것이 중요하다.

## 7. 보조 분석: 첫 7일 재방문 일수별 D30 리텐션

앞선 분석에서는 재방문이 첫 주 후반까지 이어졌는지를 확인했다.      
이 분석에서는 **첫 7일 동안 며칠이나 앱을 다시 실행했는지**에 따라 D30 리텐션이 달라지는지 확인한다.

| 그룹 | 기준 |
|---|---|
| 0일 | D1~D7 사이 앱실행 없음 |
| 1일 | D1~D7 사이 앱실행일 1일 |
| 2~3일 | D1~D7 사이 앱실행일 2~3일 |
| 4일 이상 | D1~D7 사이 앱실행일 4일 이상 |

즉, 이 분석은 단순히 첫 7일 안에 재방문했는지가 아니라, **첫 주 동안 반복적으로 앱을 실행한 사용자가 D30까지 더 잘 남는지**를 확인하기 위한 보조 분석이다.


```python
first7_revisit_day_count = (
    first7_app_launch
    .groupby("user_id")["event_date"]
    .nunique()
    .reset_index(name="first7_revisit_day_count")
)

revisit_analysis_with_count = revisit_analysis_base.merge(
    first7_revisit_day_count,
    on="user_id",
    how="left"
)

revisit_analysis_with_count["first7_revisit_day_count"] = (
    revisit_analysis_with_count["first7_revisit_day_count"]
    .fillna(0)
    .astype(int)
)

revisit_analysis_with_count["first7_revisit_day_bucket"] = np.select(
    [
        revisit_analysis_with_count["first7_revisit_day_count"] == 0,
        revisit_analysis_with_count["first7_revisit_day_count"] == 1,
        revisit_analysis_with_count["first7_revisit_day_count"].between(2, 3),
        revisit_analysis_with_count["first7_revisit_day_count"] >= 4
    ],
    [
        "0일",
        "1일",
        "2~3일",
        "4일 이상"
    ],
    default="확인 필요"
)

first7_revisit_day_summary = (
    revisit_analysis_with_count
    .groupby("first7_revisit_day_bucket")
    .agg(
        user_count=("user_id", "count"),
        d30_retained_count=("is_d30_retained", "sum")
    )
    .reset_index()
)

first7_revisit_day_summary["d30_retention_rate"] = (
    first7_revisit_day_summary["d30_retained_count"] / first7_revisit_day_summary["user_count"]
).round(4)

first7_revisit_day_summary["d30_retention_rate_percent"] = (
    first7_revisit_day_summary["d30_retention_rate"] * 100
).round(2)

first7_revisit_day_summary["sort_order"] = first7_revisit_day_summary["first7_revisit_day_bucket"].map({
    "0일": 1,
    "1일": 2,
    "2~3일": 3,
    "4일 이상": 4
})

first7_revisit_day_summary = (
    first7_revisit_day_summary
    .sort_values("sort_order")
    .drop(columns="sort_order")
)

display(first7_revisit_day_summary)
```

```python
first7_revisit_day_plot = (
    first7_revisit_day_summary
    .set_index("first7_revisit_day_bucket")
    .loc[["0일", "1일", "2~3일", "4일 이상"], "d30_retention_rate_percent"]
)

plt.figure(figsize=(8, 5))
bars = plt.bar(first7_revisit_day_plot.index, first7_revisit_day_plot.values)
plt.title("첫 7일 재방문 일수별 D30 리텐션")
plt.xlabel("첫 7일 재방문 일수")
plt.ylabel("D30 리텐션(%)")
plt.grid(axis="y", alpha=0.3)

for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f"{bar.get_height():.2f}%",
        ha="center",
        va="bottom"
    )

plt.tight_layout()
plt.show()
```

### 그래프 해석: 첫 7일 재방문 일수별 D30 리텐션

첫 7일 동안 재방문한 날짜가 0일, 1일, 2~3일인 그룹의 D30 리텐션은 모두 0.00%로 나타났다.

반면 첫 7일 동안 4일 이상 재방문한 그룹의 D30 리텐션은 37.40%로 크게 높아졌다.  
이는 첫 주 안에 단순히 한 번 다시 들어오는 것보다, **반복적으로 앱을 실행하는 초기 사용 습관**이 D30 잔존과 더 강하게 연결된다는 것을 보여준다.

따라서 개선 전략은 “7일 안에 한 번 방문”보다 “첫 7일 중 여러 날에 걸쳐 반복 방문”을 만드는 방향으로 설계하는 것이 더 적절하다.

첫 7일 이내 재방문 여부에 따른 차이가 특정 가입월에만 나타나는지, 여러 가입월에서 반복되는지 확인한다.

따라서 이 차이는 특정 월의 일시적인 현상이라기보다, 초기 재방문 행동과 장기 잔존이 연결되어 있음을 보여주는 보조 근거로 볼 수 있다.


## 9. 보조 분석: 첫 7일 재방문 실패 유저 특성

앞에서는 첫 7일 이내 재방문 여부가 D30 리텐션과 연결되는지 확인했다.  
이번에는 한 단계 더 들어가서 **어떤 사용자군에서 첫 7일 재방문 실패가 많이 발생하는지** 확인한다.

확인 기준은 아래와 같다.

| 구분 | 확인 내용 |
|---|---|
| 가입경로 | 오가닉/퍼포먼스광고/Unknown별 첫 7일 재방문 실패율 |
| 기기 | iOS/Android/Unknown별 첫 7일 재방문 실패율 |
| 알림 동의 여부 | 알림 동의/미동의별 첫 7일 재방문 실패율 |
| 온보딩 완료 여부 | 온보딩 완료/미완료별 첫 7일 재방문 실패율 |

```python
first7_failure_base = revisit_analysis_base.merge(
    user_profile[[
        "user_id",
        "signup_channel",
        "device",
        "notification_agreed",
        "is_onboarding_completed"
    ]],
    on="user_id",
    how="left"
).copy()

first7_failure_base["is_first7_revisit_failed"] = ~first7_failure_base["is_first7_revisited"]

first7_failure_base["notification_agreed_group"] = first7_failure_base["notification_agreed"].replace({
    "True": "알림 동의",
    "False": "알림 미동의",
    "Unknown": "알림 동의 Unknown",
    True: "알림 동의",
    False: "알림 미동의"
})

first7_failure_base["onboarding_group"] = np.where(
    first7_failure_base["is_onboarding_completed"],
    "온보딩 완료",
    "온보딩 미완료"
)

channel_failure_summary = (
    first7_failure_base
    .groupby("signup_channel")
    .agg(
        user_count=("user_id", "count"),
        first7_revisit_failed_count=("is_first7_revisit_failed", "sum"),
        d30_retained_count=("is_d30_retained", "sum")
    )
    .reset_index()
    .rename(columns={"signup_channel": "segment_value"})
)
channel_failure_summary["segment_type"] = "가입경로"

device_failure_summary = (
    first7_failure_base
    .groupby("device")
    .agg(
        user_count=("user_id", "count"),
        first7_revisit_failed_count=("is_first7_revisit_failed", "sum"),
        d30_retained_count=("is_d30_retained", "sum")
    )
    .reset_index()
    .rename(columns={"device": "segment_value"})
)
device_failure_summary["segment_type"] = "기기"

notification_failure_summary = (
    first7_failure_base
    .groupby("notification_agreed_group")
    .agg(
        user_count=("user_id", "count"),
        first7_revisit_failed_count=("is_first7_revisit_failed", "sum"),
        d30_retained_count=("is_d30_retained", "sum")
    )
    .reset_index()
    .rename(columns={"notification_agreed_group": "segment_value"})
)
notification_failure_summary["segment_type"] = "알림동의"

onboarding_failure_summary = (
    first7_failure_base
    .groupby("onboarding_group")
    .agg(
        user_count=("user_id", "count"),
        first7_revisit_failed_count=("is_first7_revisit_failed", "sum"),
        d30_retained_count=("is_d30_retained", "sum")
    )
    .reset_index()
    .rename(columns={"onboarding_group": "segment_value"})
)
onboarding_failure_summary["segment_type"] = "온보딩"

segment_failure_summary = pd.concat(
    [
        channel_failure_summary,
        device_failure_summary,
        notification_failure_summary,
        onboarding_failure_summary
    ],
    ignore_index=True
)

segment_failure_summary = segment_failure_summary[[
    "segment_type",
    "segment_value",
    "user_count",
    "first7_revisit_failed_count",
    "d30_retained_count"
]]

segment_failure_summary["first7_revisit_failure_rate_percent"] = (
    segment_failure_summary["first7_revisit_failed_count"] / segment_failure_summary["user_count"] * 100
).round(2)

segment_failure_summary["d30_retention_rate_percent"] = (
    segment_failure_summary["d30_retained_count"] / segment_failure_summary["user_count"] * 100
).round(2)

segment_failure_summary["sort_order"] = segment_failure_summary["segment_type"].map({
    "가입경로": 1,
    "기기": 2,
    "알림동의": 3,
    "온보딩": 4
})

segment_failure_summary = (
    segment_failure_summary
    .sort_values(["sort_order", "segment_value"])
    .drop(columns="sort_order")
)
```

```python
segment_failure_summary_output = segment_failure_summary.rename(columns={
    "segment_type": "구분",
    "segment_value": "세부 그룹",
    "user_count": "사용자 수",
    "first7_revisit_failed_count": "첫 7일 재방문 실패 수",
    "d30_retained_count": "D30 유지 수",
    "first7_revisit_failure_rate_percent": "첫 7일 재방문 실패율(%)",
    "d30_retention_rate_percent": "D30 리텐션(%)"
})

display(segment_failure_summary_output)
```

```python
segment_failure_plot = segment_failure_summary.copy()

segment_failure_plot["segment_label"] = (
    segment_failure_plot["segment_type"]
    + " | "
    + segment_failure_plot["segment_value"].astype(str)
)

segment_failure_plot = segment_failure_plot.sort_values(
    "first7_revisit_failure_rate_percent",
    ascending=True
)

plt.figure(figsize=(9, 6))
bars = plt.barh(
    segment_failure_plot["segment_label"],
    segment_failure_plot["first7_revisit_failure_rate_percent"]
)

plt.title("사용자 특성별 첫 7일 재방문 실패율")
plt.xlabel("첫 7일 재방문 실패율(%)")
plt.ylabel("사용자 특성")
plt.grid(axis="x", alpha=0.3)

for bar in bars:
    plt.text(
        bar.get_width(),
        bar.get_y() + bar.get_height() / 2,
        f"{bar.get_width():.2f}%",
        va="center",
        ha="left"
    )

plt.tight_layout()
plt.show()
```

### 그래프 해석: 사용자 특성별 첫 7일 재방문 실패율

가입경로, 기기, 알림 동의 여부에서는 첫 7일 재방문 실패율 차이가 크지 않았다.  
오가닉과 퍼포먼스광고, iOS와 Android, 알림 동의와 미동의 그룹 모두 대체로 13~16%대 수준에서 나타난다.

반면 온보딩 완료 여부에서는 차이가 더 뚜렷하다.  
온보딩 미완료 사용자의 첫 7일 재방문 실패율은 16.59%로, 온보딩 완료 사용자 13.10%보다 높다.  
D30 리텐션도 온보딩 완료 사용자가 30.75%, 미완료 사용자가 20.81%로 차이가 난다.

따라서 첫 7일 재방문 실패는 특정 가입경로나 기기 문제라기보다, **초기 온보딩 경험을 완료하지 못한 사용자에게서 상대적으로 더 많이 발생하는 문제**로 해석할 수 있다.
